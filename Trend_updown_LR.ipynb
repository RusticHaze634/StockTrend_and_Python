```python
import pandas as pd
import yfinance as yf
import statsmodels.api as sm
```

```python
data = yf.download("SKY", start = '2001-01-03' , end ='2022-05-21')
```

Creating a for loop   
5 additional columns  
shift particular lag  

```python
df['Volume'] = data.Volume.shift(1).values/1000_000_000
```

```python
for i in range(1,6):
  df ['Lag '+str(i)] = df['Today'].shift(i)

df
```

```python
df = df.reset_index()
```

```python
#df['Volume'] = data.Volume.shift(1).values/1000_000_000
```

```python
df
```

```python
df.shape
```

```python
df = df.dropna()
df
```

direction column missing
- in sklearn : up if i>0 else down for i in df['Today']

```python
df['Direction'] = [1 if i>0 else 0 for i in df['Today']]
df
```

add a constant to dataframe in statsmodes, otherwise regression will not be containing any intercept

```python
df = sm.add_constant(df)
df
```

Use the lags and the volume to predict the direction.  So here ,   
independant varibles: lags , volume;  
dependant variable :  direction  

```python
X = df[['const','Lag 1', 'Lag 2', 'Lag 3', 'Lag 4', 'Lag 5', 'Volume']]
y = df.Direction
```

```python
model = sm.Logit(y,X)
result = model.fit()
```

```python
result.summary()
```

```python
prediction = result.predict(X)
prediction
```

values > 0.5 = up  
vaues < 0.5 = down  

a confusion matrix : to see this prediction is good or bad  
- comparing the actual values vs predicted values

```python
def confusion_matrix(act, pred):
  predtrans = ['Up' if i > 0.5 else "Down" for i in pred]
  actuals = ['Up' if i > 0 else "Down" for i in act]
  confusion_matrix = pd.crosstab(pd.Series(actuals),
                                 pd.Series(predtrans),
                                 rownames = ['Actuals'],
                                 colnames=['Predicted'])
  return confusion_matrix
```

```python
confusion_matrix(y, prediction)
```

- in above table : diagonal is containing rightly predicted values 

- check : how often the model has corectly predicted   
= correctly predicted values / sum of all the values in table

```python
len(df)
```

```python
P = (2171+589)/len(df)
print("correct prediction = ", format(P, ".4f"))
```

- to check it : use test dataset now

```python
x_train = df[df.Date.dt.year < 2002][['const','Lag 1', 'Lag 2']] ##'Lag 3', 'Lag 4', 'Lag 5', 'Volume']]
y_train = df[df.Date.dt.year < 2002]['Direction']
x_test = df[df.Date.dt.year == 2002][['const','Lag 1', 'Lag 2']] ##'Lag 3', 'Lag 4', 'Lag 5', 'Volume']]
y_test = df[df.Date.dt.year == 2002]['Direction']
```

```python
model = sm.Logit(y_train,x_train)
```

```python
result = model.fit()
```

```python
prediction = result.predict(x_test)
confusion_matrix(y_test, prediction)
```

```python
P = (21+95)/len(x_test)
print("correct prediction = ", format(P, ".4f"))
```
